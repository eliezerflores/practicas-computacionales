Exercise 36 

$$
\vec{f}(x,y) = xi -yj
$$

```
import numpy as np 
import matplotlib.pyplot as plt
```

```
x_range = y_range = np.linspace(0,10,10)
x, y = np.meshgrid(x_range, y_range)
fx, fy = x, -y
```

```
plt.figure(figsize = (8,8))
plt.quiver(x,y,fx,fy)
plt.axis("scaled")
plt.show()
```

```
def trajectory_35(t):

    x = np.exp(t)
    y = np.exp(-t)

    return [x, y]
```

```
t = np.linspace(-100,100,1000)
xt, yt = trajectory_35(t)
```

```
plt.figure(figsize = (8,8))
plt.quiver(x,y,fx,fy)
plt.axis("scaled")
plt.plot(xt,yt, "b--")
plt.scatter(1 ,1,c="r",s = 100)
plt.show()
```

Exercise 36

$$
\vec{f}(x,y) = i + xj
$$

```
x_range = y_range = np.linspace(0,10,10)
x, y = np.meshgrid(x_range, y_range)
fx, fy = 1, x
```

```
def trajectory_36(t):

    x = t
    y = t**2

    return [x, y]
```

```
t = np.linspace(-100,100,1000)
xt, yt = trajectory_36(t)
```

```
plt.figure(figsize = (8,8))
plt.quiver(x,y,fx,fy)
plt.axis("scaled")
plt.plot(xt,yt, "b--")
plt.scatter(0 ,0,c="r",s = 100)
plt.show()
```
